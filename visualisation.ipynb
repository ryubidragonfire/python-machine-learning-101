## Visualisation


```python
import pandas as pd
df = pd.read_csv("./data/iris-with-header.tsv", delimiter='\t' )
df.head()
```

## Boxplot

```python
%matplotlib inline
boxplots = df.boxplot(return_type='axes')
```

**redline** = median

**side of box** = 25%, 75% percentile

**whiskers** = 1.5 times IQR

**outside of whiskers** = > 1.5 times IQR

See [Box Plot](https://en.wikipedia.org/wiki/Box_plot), and [Interquartile Range(IQR)](https://en.wikipedia.org/wiki/Interquartile_range)

```python
boxplots_by_group = df.boxplot(by='Species', return_type='axes')
```

```python
boxplots_by_group = df.boxplot(column='PetalLengthCm', by='Species', return_type='axes')
```

## Parallel Coordinates

```python
from pandas.tools.plotting import parallel_coordinates
pc = parallel_coordinates(df, 'Species')
```

## Histogram

```python
df.plot.hist()
```

## Density Distribution

```python
df.plot.density()
```

## Scatter Plot
